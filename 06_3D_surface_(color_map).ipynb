## **Ploting Functions** 

### __Sine function__

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Create the domain 
x = np.arange(0, 2*np.pi, 0.1) # start at 0, go to 2pi, go by 0.1 step
f = np.sin(x)

# NOTE: x's is a list and f's is a list
plt.plot(x, f, 'r-') # Match x's to y's and plot each point
plt.grid()     # Adding the grid function
plt.legend(['The sine Function'])
plt.xlabel('x')
plt.ylabel('f(x) = sin(x) on 0 to 2pi')
plt.title('Sine Func')
plt.show()
```

### __Sine Cosine and their sum__

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Create the domain 
x = np.arange(0, 2*np.pi, 0.1) # start at 0, go to 2pi, go by 0.1 step
f = np.sin(x)
g = np.cos(x)
h = f + g 

# plot
plt.plot(x, f, 'r.')
plt.plot(x, g, 'b-.') 
plt.plot(x, h, 'g-')  
plt.grid()     
plt.legend(['The sine Func', 'The cosine Func', 'Sum of Sine and cosine'])
plt.xlabel('x')
plt.ylabel('f(x) = sin(x) on 0 to 2pi')
plt.title('Sine Func , Cosine Funce and Their Sum')
plt.show()
```

## __Layering the plots with loops__

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Create the domain 
x = np.linspace(0, 1, 100)

legend_entries = [] 

for n in range(1, 6):
    y = x**n
    plt.plot(x, y)
    legend_entries.append('x to the power of '+str(n))

plt.grid()
plt.legend(legend_entries)
plt.xlabel('linear - x')
plt.ylabel('exp func')
plt.title('Layering the plots with loops')
plt.show()
```

## __Plotting the sequence__

Let's plot the terms in the sequence: 
$${x_{n+1}}= -0.5{x_n}+1 $$

starting with the poit: ${x_0}=3$

```
import numpy as np
import matplotlib.pyplot as plt

x = 3 # initial sequence
plt.plot(0, 3, 'b.') #plot the starting point (n=0, x0=3)

for n in range(1, 21):
    x = -0.5*x + 1
    plt.plot(n, x, 'b.')

plt.grid()
plt.xlabel('n')
plt.ylabel('x_n')
plt.title('x_(n+1) = -0.5 x_n + 1')
plt.show()
```

## __Subplots__

```
import numpy as np
import matplotlib.pyplot as plt

# Define the domain 
x =  np.linspace(0, 2*np.pi, 100)

f = lambda x: np.sin(x)
g = lambda x: 2*np.sin(x)
h = lambda x: 3*np.sin(x)
k = lambda x: 4*np.sin(x)

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12,8))

ax[0,0].plot(x, f(x), 'b--')
ax[0,0].set_ylim(-5, 5)
ax[0,0].set_title('sin(x)')
ax[0,0].grid()

ax[0,1].plot(x, g(x), 'r-.')
ax[0,1].set_ylim(-5, 5)
ax[0,1].set_title('2 sin(x)')
ax[0,1].grid()

ax[1,0].plot(x, h(x), 'go')
ax[1,0].set_ylim(-5, 5)
ax[1,0].set_title('3 sin(x)')
ax[1,0].grid()

ax[1,1].plot(x, k(x), 'k-*')
ax[1,1].set_ylim(-5, 5)
ax[1,1].set_title('4 sin(x)')
ax[1,1].grid()

plt.show()
```

## __Contour Plots__

X and Y must both be 2-D with the same shape as Z (e.g. created via `numpy.meshgrid`)

```
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


# Contour Plot
f = lambda x, y: np.sin(x)*np.sin(y)
x = np.linspace(-2, 2, 50) 
y = np.linspace(-2, 2, 50)

'''[X,Y] = meshgrid(x,y) transforms the domain specified by vectors x and y  into
   arrays X and Y , which can be used to evaluate functions of two variables and 
   three-dimensional mesh/surface plots. The rows of the output array X are copies
   of the vector x ; columns of the output array Y are copies of the vector y .'''

X, Y = np.meshgrid(x,y)
F = f(X, Y)

plt.style.use('seaborn')

plt.contour(x, y, F, 15, cmap='viridis') # 20 is the depth
plt.show()
```

##__Surface Plot__

```
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm # import the color map tools

# Contour Plot
f = lambda x, y: np.sin(x)*np.sin(y)
x = np.linspace(-3, 3, 100) 
y = np.linspace(-3, 3, 100)

X, Y = np.meshgrid(x, y)
F = f(X, Y)

plt.style.use('default')
fig = plt.figure(figsize=(8, 5))
# Setting up the axis as the 3D plot
ax = fig.gca(projection='3d') # gca(get current axis)  
ax.plot_surface(X, Y, F, cmap=cm.coolwarm)

plt.show()
```

```
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

plt.style.use('seaborn')
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}, figsize=(12, 7))

# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.viridis,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter(FormatStrFormatter('{x:.02f}'))

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

```
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

#%matplotlib qt # work locally  - pip install PyQt5
%matplotlib inline
import numpy as np

plt.style.use('seaborn')
fig = plt.figure(figsize=(12, 7))
ax = fig.gca(projection='3d')

scale = 8
# Make data.
X = np.arange(-scale, scale, 0.25)
Y = np.arange(-scale, scale, 0.25)
X, Y = np.meshgrid(X, Y)
Z = X**2 + Y**2

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.viridis,
                   linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(0, 100)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# rotate the axes and update
for angle in range(0, 360):
   ax.view_init(30, 40)

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

## __Interactive Plots With matplotlib and ipywidgets__

- `ipywidgets` for slider-based animation. 
- `matplotlib.animate`



#### **1. Plot the function ${f(x)= A sin}(Bx)$ and allow the user to interact with the values `A` and `B`**


```
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive

# Create the domain 
x = np.linspace(0, 2*np.pi, 100)

# Define a func that takes `A` And `B` as input
# and Output a plot
def plotter(A, B):
    y = A*np.sin(B*x)
    plt.plot(x, y)
    plt.grid()
    plt.ylim(-3, 3)
    plt.show()
    
iplot = interactive(plotter,
                    A=(-3, 3, 0.1),
                    B=(1, 5, 0.5)) 
iplot
```

#### **2. Plot the function ${f(t)}= 3 - e^{-t/2} $ and the user to control the time, ${t}$**

```
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interactive

# Create the domain 
t = np.arange(0, 5, 0.01)

# Define a func that takes `A` And `B` as input
# and Output a plot
def plotter(timestep):
    y = 3 - np.exp(t[:timestep]/2) # Upto the timestep/2
    plt.plot(t[:timestep], y)
    plt.xlim(0, 5)
    plt.ylim(-10, 3)
    plt.grid()
    plt.show()

iplot = interactive(plotter,
                    timestep=(1, len(t)))
iplot
```
